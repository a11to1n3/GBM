# GBM 3D Quickstart — Synthetic Volume

Demonstrates `find_optimal_k_points_gbm_3D` (D-1 pillars) on synthetic 3D data.

```python
import numpy as np
import pandas as pd
from gbm.core.mapper_3d import find_optimal_k_points_gbm_3D  # adjust if import path changes

print("Attempting 3D GBM import...")
```

```python
# Create tiny synthetic 3D volume (LW=2 for speed)
lw = 2
W, H, D = 100*lw, 100, 20
x = np.linspace(-20, 20, W)
y = np.linspace(0.4, 12, D)
z = np.linspace(-70, 70, H)

X, Y, Z = np.meshgrid(x, y, z, indexing="ij")  # note ordering

# Simple 3D field with vertical plume
C = 5e-6 + 4e-6 * np.exp(-((X)**2 + (Z)**2 + (Y-6)**2) / 80)

df = pd.DataFrame({
    "X": X.ravel(),
    "Y": Y.ravel(),
    "Z": Z.ravel(),
    "Carbon": C.ravel()
})

inside = np.ones(W*H*D, dtype=bool)
gmean = float(C.mean())

print(f"Synthetic 3D volume: {W}x{H}x{D}, mean ≈ {gmean:.2e}")
```

```python
# Run 3D GBM (D-1 pillars)
try:
    loss, mean_sens, std_sens, positions = find_optimal_k_points_gbm_3D(
        df, inside, k=6, in_CO2_avg=gmean,
        cross_section="XZ", barn_LW_ratio=lw,
        epochs=6, sampling_budget=800
    )
    print(f"3D GBM loss: {loss:.2e}")
    print(f"Sensitivity: {mean_sens:.2e} ± {std_sens:.2e}")
    print(f"Positions (first 3): {positions[:3]}")
except Exception as e:
    print(f"3D GBM not yet fully wired in this build: {e}")
```
