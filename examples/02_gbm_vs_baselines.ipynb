# GMB vs Baselines — Synthetic Comparison

Quick comparison of GBM against a few baselines on synthetic data.

```python
import numpy as np
import pandas as pd
from gbm.core import find_optimal_k_points_gbm_2D
from gbm.baselines.greedy import find_optimal_k_points_greedy_2D

print("Imports successful")
```

```python
# Tiny synthetic field
W, H = 200, 100
x = np.linspace(-20, 20, W)
z = np.linspace(-10, 10, H)
X, Z = np.meshgrid(x, z, indexing="ij")
C = 5e-6 + 3e-6 * np.exp(-((X)**2 + (Z)**2) / 40)

df = pd.DataFrame({"X": X.ravel(), "Y": np.full(W*H, 3.15), "Z": Z.ravel(), "Carbon": C.ravel()})
inside = np.ones(W*H, dtype=bool)
gmean = float(C.mean())
```

```python
# Compare GBM vs Greedy
res_gbm = find_optimal_k_points_gbm_2D(df, inside, k=5, in_CO2_avg=gmean, cross_section="Z", barn_LW_ratio=2, epochs=8)
res_greedy = find_optimal_k_points_greedy_2D(df, inside, k=5, in_CO2_avg=gmean, barn_LW_ratio=2)

print("GBM loss:", res_gbm[0])
print("Greedy loss:", res_greedy[0])
```
