# GBM Quickstart — Synthetic Data

This notebook demonstrates the core Gradient-Based Mapper (GBM) API using only synthetic data.

```python
import numpy as np
import pandas as pd
from gbm.core import find_optimal_k_points_gbm_2D

print("GBM imported successfully")
```

```python
# Create a tiny synthetic 2D plane (LW=2)
W, H = 200, 100
x = np.linspace(-20, 20, W)
z = np.linspace(-10, 10, H)
X, Z = np.meshgrid(x, z, indexing="ij")

# Simple concentration field with one plume
C = 5e-6 + 3e-6 * np.exp(-((X)**2 + (Z)**2) / 50)

df = pd.DataFrame({
    "X": X.ravel(),
    "Y": np.full(W*H, 3.15),
    "Z": Z.ravel(),
    "Carbon": C.ravel()
})

inside = np.ones(W*H, dtype=bool)
global_mean = float(C.mean())

print(f"Synthetic grid: {W}x{H}, global mean ≈ {global_mean:.2e}")
```

```python
# Run GBM
loss, mean_sens, std_sens, positions = find_optimal_k_points_gbm_2D(
    df, inside, k=5, in_CO2_avg=global_mean,
    cross_section="Z", barn_LW_ratio=2,
    epochs=8, sampling_budget=500
)

print(f"Best 5-point loss: {loss:.2e}")
print(f"Sensitivity (mean ± std): {mean_sens:.2e} ± {std_sens:.2e}")
print("Selected positions (x, z):")
for p in positions:
    print(f"  {p}")
```
